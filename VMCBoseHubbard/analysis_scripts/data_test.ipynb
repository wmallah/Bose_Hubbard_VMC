```python

import numpy as np
import matplotlib.pyplot as plt
L=64
π=np.pi
y = np.array([
    -0.1625881E+01,
    -0.1048148E+01,
    -0.7777309E+00,
    -0.6105601E+00,
    -0.4930722E+00,
    -0.4043681E+00,
    -0.3347432E+00,
    -0.2786680E+00,
    -0.2327923E+00,
    -0.1949050E+00,
    -0.1633171E+00,
    -0.1366961E+00,
    -0.1143394E+00,
    -0.9545265E-01,
    -0.7940817E-01,
    -0.6579527E-01,
    -0.5437115E-01,
    -0.4466728E-01,
    -0.3659925E-01,
    -0.2988174E-01,
    -0.2420967E-01,
    -0.1944907E-01,
    -0.1543752E-01,
    -0.1201605E-01,
    -0.9099452E-02,
    -0.6679389E-02,
    -0.4549107E-02,
    -0.3009113E-02,
    -0.1798847E-02,
    -0.9387513E-03,
    -0.4239784E-03,
    -0.4166991E-04,
    -0.4166991E-04,######## extra point
])
ν= np.zeros(L)
x = np.arange(L)
ν[0]=y[0]
for i in range(1,len(y)):
    ν[i]=y[i]
    ν[L-i]=ν[i]
plt.figure(figsize=(7, 4))
plt.plot(x, ν, marker='o')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Extracted Data")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python

# Perform Discrete Fourier Transform
fft_result = np.fft.fft( ν)

# Calculate corresponding q values
q_values = 2  *np.pi*  np.arange(L) / L
ymax=int(L/2)+1
# Plot the results
plt.figure()
plt.plot(q_values[0:ymax], abs(fft_result[0:ymax])*q_values[0:ymax]**2, marker='o', linestyle='-')
plt.xlabel('q values')
plt.ylabel('FFT Magnitude')
plt.title('Discrete Fourier Transform over 64 sites')
plt.show()
```
