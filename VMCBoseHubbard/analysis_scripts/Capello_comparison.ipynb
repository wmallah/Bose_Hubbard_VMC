```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import seaborn as sns

sns.set_theme()
sns.set_style("ticks")
sns.set_context("talk")
sns.color_palette("colorblind")
```

```python
def load_capello_data(filepath):
    data = pd.read_csv(filepath).to_numpy()
    q = data[:, 0]
    vq_q2 = data[:, 1]
    return q, vq_q2

def load_vmc_vq(filepath):
    data = np.loadtxt(filepath, comments="#", ndmin=2)

    q = data[:, 0]
    vq = data[:, 1]
    vq_q2 = data[:, 2]

    return q, vq, vq_q2

def load_vmc_vr(filepath):
    data = np.loadtxt(filepath, comments="#", ndmin=2)

    r = data[:, 0].astype(int)
    vr = data[:, 1]

    return r, vr


def plot_vq_comparison(ax, q_vmc, vq_q2_vmc,
                       q_capello, vq_q2_capello,
                       U):

    ax.plot(q_capello, vq_q2_capello,
            linestyle="--",
            color="red",
            label=f"Capello U={U}")

    ax.plot(q_vmc, vq_q2_vmc,
            marker="o",
            linestyle="-",
            color="black",
            label=f"Our VMC U={U}")
```

```python
# -----------------------
# Load Capello data
# -----------------------

capello_q_U2, capello_vq_q2_U2 = load_capello_data(
    "../data/Capello_v_q_q_squared_vs_q_U_2_site_60.csv"
)

# -----------------------
# Load our VMC result
# -----------------------

vmc_q_U2, vmc_vq_U2, vmc_vq_q2_U2 = load_vmc_vq(
    "../data/C/1D/L60_N60/jastrow_realspace/VMC_vq_vs_q_UoverT2.0.dat"
)

# -----------------------
# Plot
# -----------------------

fig, ax = plt.subplots()

plot_vq_comparison(ax,
                   vmc_q_U2, vmc_vq_q2_U2,
                   capello_q_U2, capello_vq_q2_U2,
                   U=2)

ax.set_xlim(0, np.pi)
ax.set_xlabel(r"$|q|$")
ax.set_ylabel(r"$v_q q^2$")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
def expand_reduced_jastrow(vr, L, includes_R0=True, subtract_offset=False):
    """
    Expand a reduced 1D translationally invariant Jastrow potential into
    a full length-L periodic real-space kernel.

    Conventions
    -----------
    If includes_R0=True:
        vr[0] -> R = 0
        vr[1] -> R = 1
        ...
        vr[Rmax] -> R = Rmax

        length(vr) = floor(L/2) + 1

    If includes_R0=False:
        vr[0] -> R = 1
        vr[1] -> R = 2
        ...
        vr[Rmax-1] -> R = Rmax

        length(vr) = floor(L/2)

    Returns
    -------
    vfull : ndarray, shape (L,)
        Full periodic kernel v(R), R = 0, ..., L-1.
    """

    vr = np.asarray(vr, dtype=float).copy()
    Rmax = L // 2

    if includes_R0:
        expected_len = Rmax + 1
    else:
        expected_len = Rmax

    if len(vr) != expected_len:
        raise ValueError(
            f"Expected len(vr) = {expected_len}, got {len(vr)} "
            f"for L = {L}, includes_R0 = {includes_R0}"
        )

    if subtract_offset:
        vr -= vr[-1]

    vfull = np.zeros(L, dtype=float)

    if includes_R0:
        # R = 0
        vfull[0] = vr[0]

        if L % 2 == 0:
            # R = 1, ..., L/2 - 1
            for R in range(1, Rmax):
                vfull[R] = vr[R]
                vfull[L - R] = vr[R]

            # R = L/2 special point
            vfull[Rmax] = vr[Rmax]

        else:
            # R = 1, ..., floor(L/2)
            for R in range(1, Rmax + 1):
                vfull[R] = vr[R]
                vfull[L - R] = vr[R]

    else:
        # No R = 0 entry. vr[R-1] corresponds to physical distance R.

        if L % 2 == 0:
            # R = 1, ..., L/2 - 1
            for R in range(1, Rmax):
                vfull[R] = vr[R - 1]
                vfull[L - R] = vr[R - 1]

            # R = L/2 special point
            vfull[Rmax] = vr[Rmax - 1]

        else:
            # R = 1, ..., floor(L/2)
            for R in range(1, Rmax + 1):
                vfull[R] = vr[R - 1]
                vfull[L - R] = vr[R - 1]

    return vfull

import numpy as np


def realspace_to_momentum_fft(vr, L, includes_R0=True):
    """
    Convert reduced real-space Jastrow coefficients v_r to v_q using numpy FFT.

    If includes_R0=True:
        vr[0] -> R = 0
        vr[1] -> R = 1
        ...
        vr[L//2] -> R = L//2

    If includes_R0=False:
        vr[0] -> R = 1
        vr[1] -> R = 2
        ...
        vr[L//2 - 1] -> R = L//2
    """

    vr = np.asarray(vr, dtype=float)
    Rmax = L // 2

    if includes_R0:
        if len(vr) != Rmax + 1:
            raise ValueError(f"Expected len(vr) = {Rmax + 1}, got {len(vr)}")

        vfull = np.zeros(L)

        # R = 0
        vfull[0] = vr[0]

        # 1 <= R < L/2
        for R in range(1, Rmax):
            vfull[R] = vr[R]
            vfull[L - R] = vr[R]

        # R = L/2 for even L
        if L % 2 == 0:
            vfull[Rmax] = vr[Rmax]
        else:
            vfull[Rmax] = vr[Rmax]
            vfull[L - Rmax] = vr[Rmax]

    else:
        if len(vr) != Rmax:
            raise ValueError(f"Expected len(vr) = {Rmax}, got {len(vr)}")

        vfull = np.zeros(L)

        # vr[R-1] corresponds to physical distance R
        for R in range(1, Rmax):
            vfull[R] = vr[R - 1]
            vfull[L - R] = vr[R - 1]

        # R = L/2 for even L
        if L % 2 == 0:
            vfull[Rmax] = vr[Rmax - 1]
        else:
            vfull[Rmax] = vr[Rmax - 1]
            vfull[L - Rmax] = vr[Rmax - 1]

    # numpy FFT convention:
    # vq[m] = sum_R vfull[R] exp(-2π i m R / L)
    vq_full = np.fft.fft(vfull).real

    q = np.array([2.0 * np.pi * m / L for m in range(Rmax + 1)])
    vq = vq_full[:Rmax + 1]

    return q, vq
```

```python
VMC_q, VMC_vq = realspace_to_momentum_fft(VMC_vr_Ucrit, 64)
Capello_q, Capello_vq = realspace_to_momentum_fft(Capello_vr_Ucrit, 64)

plt.plot(VMC_q, abs(VMC_vq*VMC_q**2), label="Our VMC")
plt.plot(Capello_q, abs(Capello_vq*Capello_q**2), label="Capello")
plt.xlabel(r"$q$")
plt.ylabel(r"$v_qq^2$")
plt.legend(loc="best")
plt.tight_layout()
```
