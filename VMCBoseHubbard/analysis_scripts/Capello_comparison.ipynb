```python
def load_capello_data(filepath):
    data = pd.read_csv(filepath).to_numpy()
    q = data[:, 0]
    vq_q2 = data[:, 1]
    return q, vq_q2

def load_vmc_vq(filepath):
    data = np.loadtxt(filepath, comments="#", ndmin=2)

    q = data[:, 0]
    vq = data[:, 1]
    vq_q2 = data[:, 2]

    return q, vq, vq_q2

def load_vmc_vr(filepath):
    data = np.loadtxt(filepath, comments="#", ndmin=2)

    r = data[:, 0].astype(int)
    vr = data[:, 1]

    return r, vr


def plot_vq_comparison(ax, q_vmc, vq_q2_vmc,
                       q_capello, vq_q2_capello,
                       U):

    ax.plot(q_capello, vq_q2_capello,
            linestyle="--",
            color="red",
            label=f"Capello U={U}")

    ax.plot(q_vmc, vq_q2_vmc,
            marker="o",
            linestyle="-",
            color="black",
            label=f"Our VMC U={U}")
```

```python
from scipy.interpolate import interp1d


def load_capello_vq_from_vqq2(filepath):
    """
    CSV columns:
        q, v_q * q^2
    """
    data = pd.read_csv(filepath, header=None).to_numpy(dtype=float)

    q = data[:, 0]
    vq_q2 = data[:, 1]

    mask = np.abs(q) > 1e-12
    q = q[mask]
    vq_q2 = vq_q2[mask]

    vq = vq_q2 / (q**2)
    return q, vq


def realspace_to_momentum_jastrow(vr, L):
    """
    Forward transform using the same discrete convention.

    vr[0] = v_{r=0}, ..., vr[Rmax] = v_{r=Rmax}
    """
    Rmax = L // 2
    q = 2.0 * np.pi * np.arange(0, Rmax + 1) / L
    vq = np.zeros(Rmax + 1, dtype=float)

    if L % 2 == 0:
        for iq, qq in enumerate(q):
            s = vr[0]
            for r in range(1, Rmax):
                s += 2.0 * vr[r] * np.cos(qq * r)
            s += vr[Rmax] * np.cos(qq * Rmax)
            vq[iq] = s
    else:
        for iq, qq in enumerate(q):
            s = vr[0]
            for r in range(1, Rmax + 1):
                s += 2.0 * vr[r] * np.cos(qq * r)
            vq[iq] = s

    return q, vq


def inverse_capello_vq_to_vr(q_data, vq_data, L, gauge="tail_zero"):
    """
    Invert nonzero-q Capello v_q data back to real space.

    Since q=0 is missing, the real-space potential is only defined up to
    an additive constant. We first solve in the gauge v0=0, then shift.
    """
    Rmax = L // 2

    # discrete nonzero lattice q-grid
    q_target = 2.0 * np.pi * np.arange(1, Rmax + 1) / L

    # interpolate digitized Capello data onto lattice q-grid
    interp_fun = interp1d(
        q_data,
        vq_data,
        kind="linear",
        bounds_error=False,
        fill_value="extrapolate",
    )
    vq_target = interp_fun(q_target)

    # Solve A x = vq_target for x = [v1, ..., vRmax]
    A = np.zeros((Rmax, Rmax), dtype=float)

    if L % 2 == 0:
        for m, qq in enumerate(q_target):
            for r in range(1, Rmax):
                A[m, r - 1] = 2.0 * np.cos(qq * r)
            A[m, Rmax - 1] = np.cos(qq * Rmax)
    else:
        for m, qq in enumerate(q_target):
            for r in range(1, Rmax + 1):
                A[m, r - 1] = 2.0 * np.cos(qq * r)

    vr_pos = np.linalg.solve(A, vq_target)

    vr = np.zeros(Rmax + 1, dtype=float)
    vr[1:] = vr_pos

    # gauge choice
    if gauge == "v0_zero":
        pass
    elif gauge == "tail_zero":
        vr -= vr[-1]
    elif gauge == "mean_zero":
        vr -= np.mean(vr)
    else:
        raise ValueError("gauge must be 'v0_zero', 'tail_zero', or 'mean_zero'")

    r = np.arange(Rmax + 1, dtype=int)
    return r, vr, q_target, vq_target


def plot_capello_realspace_from_file(filepath, L=60, gauge="tail_zero"):
    q_capello, vq_capello = load_capello_vq_from_vqq2(filepath)

    r, vr, q_target, vq_target = inverse_capello_vq_to_vr(
        q_capello, vq_capello, L=L, gauge=gauge
    )

    # consistency check: forward transform reconstructed vr
    q_full, vq_reconstructed = realspace_to_momentum_jastrow(vr, L)

    plt.figure(figsize=(7, 5))
    plt.plot(r, vr, marker="o")
    plt.xlabel("r")
    plt.ylabel(r"$v_r$")
    plt.title(f"Capello inverse transform to real space (L={L}, gauge={gauge})")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 5))
    plt.plot(q_capello, vq_capello, "o", label="Capello digitized $v_q$")
    plt.plot(q_target, vq_target, "-", label="Interpolated to lattice q")
    plt.plot(q_full[1:], vq_reconstructed[1:], "--", label="Reconstructed from $v_r$")
    plt.xlabel("q")
    plt.ylabel(r"$v_q$")
    plt.title("Momentum-space consistency check")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    return r, vr
```

```python

def realspace_to_momentum_jastrow_exclude_q0(vr, L, subtract_tail=False, tail_size=5):
    """
    Transform reduced real-space Jastrow coefficients vr[R]
    into the momentum-space kernel V_q corresponding to

        log Psi = -(1/(2L)) * sum_q V_q n_q n_{-q}

    excluding q = 0 from the returned arrays.

    Parameters
    ----------
    vr : array-like
        Real-space Jastrow coefficients with vr[0] = R=0, ..., vr[Rmax] = Rmax.
        Must have length Rmax + 1 where Rmax = L//2.
    L : int
        System size.
    subtract_tail : bool, optional
        If True, subtract the mean of the last `tail_size` entries of vr before transforming.
    tail_size : int, optional
        Number of tail points used for mean subtraction.

    Returns
    -------
    q_vals : np.ndarray
        Nonzero momentum values q = 2*pi*m/L for m = 1, ..., Rmax.
    Vq : np.ndarray
        Momentum-space Jastrow kernel evaluated at those q values.
    """
    vr = np.asarray(vr, dtype=float)
    Rmax = L // 2

    if len(vr) != Rmax + 1:
        raise ValueError(f"Expected len(vr) = {Rmax + 1}, got {len(vr)}")

    vr_work = vr.copy()

    if subtract_tail:
        n_tail = min(tail_size, len(vr_work))
        tail_mean = np.mean(vr_work[-n_tail:])
        vr_work -= tail_mean

    # Exclude q=0 by starting m at 1
    m_vals = np.arange(1, Rmax + 1)
    q_vals = 2.0 * np.pi * m_vals / L
    Vq = np.zeros_like(q_vals, dtype=float)

    if L % 2 == 0:
        for iq, q in enumerate(q_vals):
            s = 2.0 * vr_work[0]  # R = 0

            for R in range(1, Rmax):
                s += 4.0 * vr_work[R] * np.cos(q * R)

            s += 2.0 * vr_work[Rmax] * np.cos(q * Rmax)  # R = L/2 special term
            Vq[iq] = s
    else:
        for iq, q in enumerate(q_vals):
            s = 2.0 * vr_work[0]  # R = 0

            for R in range(1, Rmax + 1):
                s += 4.0 * vr_work[R] * np.cos(q * R)

            Vq[iq] = s

    return q_vals, Vq
```

# Plot the difference between my parameters and Capello's parameters

```python
from scipy.interpolate import interp1d

def interpolate_capello_to_vmc(q_vmc, q_capello, vq_q2_capello):
    interp = interp1d(
        q_capello,
        vq_q2_capello,
        kind='linear',
        bounds_error=False,
        fill_value="extrapolate"
    )
    return interp(q_vmc)


def compare_vq_q2(q_vmc, vq_vmc, q_capello, vq_q2_capello):
    """
    Compare VMC and Capello in v_q * q^2 space.

    Inputs:
        q_vmc            : VMC q grid
        vq_vmc           : VMC v_q
        q_capello        : Capello q grid
        vq_q2_capello    : Capello v_q * q^2 (digitized)

    Returns:
        q_common
        vmc_vq_q2
        capello_vq_q2_interp
        delta
    """

    # Compute VMC v_q * q^2
    vq_q2_vmc = 0.5*vq_vmc * q_vmc**2

    # Restrict to overlapping q-range
    qmin = max(q_vmc.min(), q_capello.min())
    qmax = min(q_vmc.max(), q_capello.max())

    mask = (q_vmc >= qmin) & (q_vmc <= qmax)
    q_common = q_vmc[mask]
    vq_q2_vmc_common = vq_q2_vmc[mask]

    # Interpolate Capello onto VMC grid
    vq_q2_cap_interp = interpolate_capello_to_vmc(
        q_common, q_capello, vq_q2_capello
    )

    # Difference
    delta = vq_q2_cap_interp - vq_q2_vmc_common

    return q_common, vq_q2_vmc_common, vq_q2_cap_interp, delta
```

```python
q_common, vmc_q2, cap_q2, delta = compare_vq_q2(
    VMC_q_U4, VMC_vq_U4,
    Capello_q_U4, Capello_vq_q2_U4
)

plt.plot(q_common, vmc_q2, label="VMC")
plt.plot(q_common, cap_q2, label="Capello (interp)")
plt.legend()
plt.xlabel("q")
plt.ylabel(r"$v_q q^2$")
plt.show()

plt.plot(q_common, delta, 'o-')
plt.axhline(np.mean(delta), linestyle='--', label="mean offset")
plt.legend()
plt.xlabel("q")
plt.ylabel("Capello - VMC")
plt.show()
```
